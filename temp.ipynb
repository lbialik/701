```python
from pprint import pprint
from utils.data_utils import *
import copy
import json
```

```python
data = json.load(open("data/augmented_data.json"))

for condition in data:
    print('condition: ', condition)
    for item in data[condition]:
        print('item: ', item)
        print('example sentence: ', data[condition][item]['example_sentence'])
        print("NP")
        for measure in data[condition][item]['NP']:
            print('measure: ', measure, ' = ', data[condition][item]['NP'][measure])
        print("verb")
        for measure in data[condition][item]['verb']:
            print('measure: ', measure, ' = ', data[condition][item]['verb'][measure])
        print()
```

```python
def split_sentence_on(sentence, index):
    split_sentence = sentence.split('^')
    beginning, middle, end = ''.join(split_sentence[:index]).strip(), split_sentence[index].strip(), ''.join(split_sentence[index+1:]).strip()
    return beginning, middle, end

def run_model(lang_model, segments):
    print(f'lang model = {lang_model}')
    print(f'segments = {segments}')
    return 1

def add_LM_measures(lang_model):
    new_data = copy.deepcopy(data)
    for condition in data:
        # print('condition: ', condition)
        for item in data[condition]:
            # print('item: ', item)
            sentence = data[condition][item]['example_sentence']
                np_beginning, np_middle, np_end = split_sentence_on(sentence, NP_region(condition))
                v_beginning, v_middle, v_end = split_sentence_on(sentence, verb_region(condition))
                new_data[condition][item]['measurements'][lang_model+'_NP'] = run_model(lang_model, [np_beginning, np_middle])
                new_data[condition][item]['measurements'][lang_model+'_verb'] = run_model(lang_model, [v_beginning, v_middle])
    return new_data

data = add_LM_measures('GPT2')
data = add_LM_measures('LSTM')
pprint(data)
```
