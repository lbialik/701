```python
from pprint import pprint
from utils.data_utils import *
import copy
import json
```

```python
data = json.load(open("data/augmented_data.json"))

for condition in data:
    print('condition: ', condition)
    for item in data[condition]:
        print('item: ', item)
        print('example sentence: ', data[condition][item]['example_sentence'])
        print("NP")
        for measure in data[condition][item]['NP']:
            print(measure, ' = ', data[condition][item]['NP'][measure])
        print("verb")
        for measure in data[condition][item]['verb']:
            print(measure, ' = ', data[condition][item]['verb'][measure])
        print()
```

```python
data = json.load(open("data/augmented_data.json"))

for item in data['ORC']:
    diff = data['ORC'][item]['verb']['GPT2'] - data['SRC'][item]['verb']['GPT2']
    if diff > 0:
        print(f'item {item}')
        print('orc = ', data['ORC'][item]['verb']['GPT2'])
        print('src = ', data['SRC'][item]['verb']['GPT2'])
        print('example ORC sentence: ', data['ORC'][item]['example_sentence'])
        print('example SRC sentence: ', data['SRC'][item]['example_sentence'])
        print(f'diff = {diff}')
```
